```python
import numpy as np
import matplotlib.pyplot as plt

x = [10, 20, 30, 40, 50]
y1 = [1, 10, 20, 30, 50]
y2 = [0, 40, 20, 60, 80]
y3 = [1, 30, 50, 70, 90]

y = np.vstack([y1, y2, y3])

labels = ["Abusers with treatment recieved ", "Abusers without treatment", "Abusers"]

fig, ax = plt.subplots()
ax.stackplot(x, y1, y2, y3, labels=labels)
ax.legend(loc='upper left')
plt.show()

fig, ax = plt.subplots()
ax.stackplot(x, y)
plt.show()
```
